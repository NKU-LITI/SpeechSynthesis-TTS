#### falc to wav

```python
## falc转wav
import os
from pydub import AudioSegment

# 输入和输出目录
input_dir = r"D:\Users\86136\NKU\homework\语音信息处理\final\Fine-Tune\p237\p237-flac"
output_dir = r"D:\Users\86136\NKU\homework\语音信息处理\final\Fine-Tune\p237\p237-wav"

# 如果输出目录不存在，创建它
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# 遍历输入目录中的所有文件
for filename in os.listdir(input_dir):
    if filename.endswith(".flac"):
        # 构建完整的文件路径
        flac_path = os.path.join(input_dir, filename)
        wav_filename = filename.replace(".flac", ".wav")
        wav_path = os.path.join(output_dir, wav_filename)

        # 加载 FLAC 文件并转换为 WAV
        audio = AudioSegment.from_file(flac_path, format="flac")
        audio.export(wav_path, format="wav")

        print(f"转换 {filename} 为 {wav_filename}")

print("所有文件已转换完成。")
```

### wav file settle

```python
## 删除带有mic2的文件，并将带有mic1的文件重命名为去掉_mic1的版本

import os

# 指定目标目录
directory = r'D:\Users\86136\NKU\homework\语音信息处理\final\Fine-Tune\p237\p237-wav'

# 遍历目录中的文件
for filename in os.listdir(directory):
    # 构建文件的完整路径
    file_path = os.path.join(directory, filename)

    # 检查文件名是否包含 mic2，并删除这些文件
    if '_mic2' in filename:
        os.remove(file_path)
        print(f"Deleted {file_path}")

    # 检查文件名是否包含 mic1，并重命名这些文件
    elif '_mic1' in filename:
        new_filename = filename.replace('_mic1', '')
        new_file_path = os.path.join(directory, new_filename)
        os.rename(file_path, new_file_path)
        print(f"Renamed {file_path} to {new_file_path}")

print("Task completed!")
```

## txt generate

```python
## 读取所有的文本文件整理到一个txt中，格式: VCTK/p239/p239-wav/<txt的文件名>.wav|<txt>


# 输入和输出目录
input_dir = r"D:\Users\86136\NKU\homework\语音信息处理\final\Fine-Tune\p237\p237-txt"
output_file = r"D:\Users\86136\NKU\homework\语音信息处理\final\Fine-Tune\p237\p237.txt"

# 打开输出文件
with open(output_file, 'w', encoding='utf-8') as outfile:
    # 遍历输入目录中的所有文件
    for filename in os.listdir(input_dir):
        if filename.endswith(".txt"):
            # 构建完整的文件路径
            txt_path = os.path.join(input_dir, filename)
            # 读取文本文件内容
            with open(txt_path, 'r', encoding='utf-8') as infile:
                text = infile.read().strip()
            # 获取文件名（不包含扩展名）
            base_filename = os.path.splitext(filename)[0]
            # 写入到输出文件
            outfile.write(f"VCTK/p237/p237-wav/{base_filename}.wav|{text}\n")

print("所有文件已整理完成。")
```

## train set and valid set

```python
## 划分训练集和验证集

import random

# 原始文件路径
input_file = r"D:\Users\86136\NKU\homework\语音信息处理\final\Fine-Tune\p237\p237.txt"

# 划分后的训练集和验证集文件路径
train_file = r"D:\Users\86136\NKU\homework\语音信息处理\final\Fine-Tune\p237\vctk_p239_audio_text_train_filelist.txt"
val_file = r"D:\Users\86136\NKU\homework\语音信息处理\final\Fine-Tune\p237\vctk_p239_audio_text_val_filelist.txt"

# 读取原始文件内容
with open(input_file, 'r', encoding='utf-8') as f:
    lines = f.readlines()

# 打乱数据顺序
random.shuffle(lines)

# 划分训练集和验证集的比例
train_ratio = 0.8  # 训练集比例
num_total = len(lines)
num_train = int(train_ratio * num_total)

# 分割训练集和验证集
train_data = lines[:num_train]
val_data = lines[num_train:]

# 写入训练集文件
with open(train_file, 'w', encoding='utf-8') as f_train:
    for line in train_data:
        f_train.write(line)

# 写入验证集文件
with open(val_file, 'w', encoding='utf-8') as f_val:
    for line in val_data:
        f_val.write(line)

print(f"划分完成，训练集文件保存为：{train_file}")
print(f"划分完成，验证集文件保存为：{val_file}")
```
